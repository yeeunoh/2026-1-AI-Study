빙어데이터 14마리 데이터값

```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7,
29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0,
32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0,
34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0,
38.5, 38.5, 39.5, 41.0, 41.0]
```

```python
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0,
450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0,
650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0,
720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0,
975.0, 950.0]
```

```python
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3,
11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0,
                9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
#새로운 데이터 추가
new_fish = [
    [30, 600],
    [12, 10],
    [25, 150]
]
```

```python
plt.scatter(bream_length, bream_weight, label='bream')
plt.scatter(smelt_length, smelt_weight, label='smelt')

for i, fish in enumerate(new_fish):
  plt.scatter(fish[0], fish[1], marker='^', label='new fish')

plt.xlabel('length')
plt.ylabel('weight')
plt.legend(loc='upper left')
plt.show()
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

print(length)
print(weight)
```

```python
fish_data = [[l,w] for l,w in zip(length, weight)]
#zip을 통행 [길이,무게]로 데이터 형식 바꾸기 -> scikit-lear 입력 데이터를 2차원 배열로 받는다

print(fish_data)
```

```python
fish_target = [1]*35 + [0]*14
#도미 35마리, 빙어 14마리

print(fish_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier() #k-최근접 이웃 알고리즘으로 모델 학습
```

K-최근접 이웃 알고리즘 (K-Nearest Neighbors)

: 이 데이터와 거리가 가까운 데이터 k개를 찾아서 그 이웃들이 어떤 정답을 가지고 있는지 보고 다수결로 판단하자

```python
kn.fit(fish_data, fish_target)

#fish_data 입력데이터와 fish_target 정답데이터를 이용해
#이 물고기들의 길이와 무게를 보고, 도미/빙어가 어떤건지 기억해
```

```python
kn.score(fish_data, fish_target)

#모델의 정확도를 계산
#fish_data 에 있는 물고기 -> 도미/빙어 예측 -> 예측 결과를 실제 정답인 fish_target과 비교 -> 맞힌 비율 계산
```

정확도 100% -> 데이터를 정확하게 하나도 틀리지 않고 맞췄다 ?

-> 모델이 이미 봤던 문제를 다시 푸는 문제 발생

훈련에 사용한 데이터를 그대로 다시 평가해 성능이 높음

-> 학습데이터와 평가데이터를 나누자 !

```python
kn.predict(new_fish)
```
